```python
# 1) Import pandas, numpy
import pandas as pd
import numpy as np
```

```python
# 2) Load dataset using pd.read_csv()
url = "https://raw.githubusercontent.com/laxmimerit/All-CSV-ML-Data-Files-Download/master/IMDB-Movie-Data.csv"
df = pd.read_csv(url)
```

### 4) Check and remove duplicates

```python
# 4) Check and remove duplicates
df = df.drop_duplicates()
```

```python
# 5) Rename columns ("Title" > "Movie_Title")
df.rename(columns={"Title": "Movie_Title"}, inplace=True)
```

```python
# 6) Get unique genres
unique_genres = df['Genre'].unique()
unique_genres
```

```python
# 7) Extract number of genres per movie
df['Num_Genres'] = df['Genre'].apply(lambda x: len(x.split(',')))
df['Num_Genres']
```

```python
# 8) Handle missing values in 'Revenue (Millions)' and 'Metascore'
df['Revenue (Millions)'] = df['Revenue (Millions)'].fillna(df['Revenue (Millions)'].median())
df['Metascore'] = df['Metascore'].fillna(df['Metascore'].median())
df['Revenue (Millions)']
```

```python
df['Metascore']
```

```python
# 9) Top 5 movies by Rating
top5_movies = df.nlargest(5, 'Rating')[['Movie_Title', 'Rating']]
top5_movies
```

```python
# 10) Movies with highest/lowest revenue
highest_rev = df.loc[df['Revenue (Millions)'].idxmax()]
lowest_rev = df.loc[df['Revenue (Millions)'].idxmin()]
highest_rev
```

```python
lowest_rev
```

```python
# 11) Avg. rating by Director
avg_rating_director = df.groupby('Director')['Rating'].mean().sort_values(ascending=False)
avg_rating_director
```

```python
# 12) Movies per year
movies_per_year = df['Year'].value_counts().sort_index()
movies_per_year
```

```python
# 13) Movies with rating > 8 & revenue > 100M
high_quality = df[(df['Rating'] > 8) & (df['Revenue (Millions)'] > 100)]
print("\nMovies with Rating > 8 & Revenue > 100M:\n", high_quality[['Movie_Title','Rating','Revenue (Millions)']])
```

```python
# 14) Group by Director: avg Rating, Revenue
director_group = df.groupby('Director').agg({'Rating':'mean','Revenue (Millions)':'mean'}).sort_values(by='Rating',ascending=False)
director_group
```

```python
# 15) Group by Year: count, avg Rating
year_group = df.groupby('Year').agg({'Movie_Title':'count','Rating':'mean'})
year_group
```

```python
# 16) Create new column Profitability = Revenue / Duration
df['Profitability'] = df['Revenue (Millions)'] / df['Runtime (Minutes)']
df['Profitability']
```

```python
# a) Find top 10 most profitable movies
top10_profitable = df.nlargest(10, 'Profitability')[['Movie_Title','Profitability']]
top10_profitable
```

```python
# 17) Create column categorizing movies by rating
def categorize_rating(r):
    if r < 6:
        return "Poor"
    elif 6 <= r < 7.5:
        return "Average"
    else:
        return "Excellent"

df['Rating_Category'] = df['Rating'].apply(categorize_rating)
df['Rating_Category']
```

```python
# 18) Custom function to score movies
def movie_score(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']

df['High_Score'] = df.apply(movie_score, axis=1)

# Display final dataset preview
df.head()
```
